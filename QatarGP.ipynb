```python

import fastf1
import os

# Create cache directory
cache_dir = '/content/fastf1_cache'
os.makedirs(cache_dir, exist_ok=True)

# Enable FastF1 caching
fastf1.Cache.enable_cache(cache_dir)
```

```python
import pandas as pd

def load_race_features(year, round_no):
    try:
        session_q = fastf1.get_session(year, round_no, 'Q')
        session_r = fastf1.get_session(year, round_no, 'R')
        session_q.load()
        session_r.load()
    except Exception as e:
        print(f"Skipping Year {year} Round {round_no}: {e}")
        return None

    # Qualifying results
    q_results = session_q.results[['DriverNumber', 'Abbreviation', 'TeamName', 'Position']].copy()
    q_results.rename(columns={'Position':'GridPos'}, inplace=True)

    # Race results
    r_results = session_r.results[['DriverNumber','Position','Status','Points']].copy()
    r_results.rename(columns={'Position':'FinishPos'}, inplace=True)

    # Merge
    df = q_results.merge(r_results, on='DriverNumber')
    df['Year'] = year
    df['Round'] = round_no

    # Target variable: did driver win the race?
    df['TargetWin'] = (df['FinishPos'] == 1).astype(int)

    return df
```

```python
years_to_try = [2019, 2020, 2021, 2022, 2023, 2024]
races = []

for year in years_to_try:
    for rnd in range(1, 25):  # max possible rounds
        df_race = load_race_features(year, rnd)
        if df_race is not None:
            races.append(df_race)

# Combine all races
if races:
    df = pd.concat(races, ignore_index=True)
else:
    raise ValueError("No race data loaded. Check cache or FastF1 session availability.")

# Quick check
df.head()
```

```python
df['GridPos'] = pd.to_numeric(df['GridPos'], errors='coerce')
df['FinishPos'] = pd.to_numeric(df['FinishPos'], errors='coerce')

# Features
features = ['GridPos']  # can expand later (Team, Historical Avg, etc.)
target = 'TargetWin'
groups = df['Round']  # for GroupKFold CV

X = df[features]
y = df[target]
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GroupKFold
from sklearn.metrics import log_loss
import numpy as np

model = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    n_estimators=500,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.7,
    use_label_encoder=False
)

gkf = GroupKFold(n_splits=5)
scores = []

for train_idx, val_idx in gkf.split(X, y, groups):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # XGBClassifier.fit no longer accepts early_stopping_rounds in some versions
    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val)],
        verbose=False
    )

    preds = model.predict_proba(X_val)[:,1]
    scores.append(log_loss(y_val, preds))

print("Mean CV logloss:", np.mean(scores))
```

```python
qatar_qualifying = pd.DataFrame({
    'DriverNumber': [4, 20, 1, 63, 22, 23, 16, 44, 10, 55, 14, 77],
    'DriverName': ["Lando Norris", "Oscar Piastri", "Max Verstappen", "George Russell",
                   "Yuki Tsunoda", "Alexander Albon", "Charles Leclerc", "Lewis Hamilton",
                   "Pierre Gasly", "Carlos Sainz", "Fernando Alonso", "Lance Stroll"],
    'GridPos': [3, 4, 1, 2, 6, 8, 5, 7, 9, 10, 11, 12]
})

qatar_qualifying['WinProb'] = model.predict_proba(qatar_qualifying[['GridPos']])[:,1]

# Sort by probability
qatar_qualifying = qatar_qualifying.sort_values(by='WinProb', ascending=False).reset_index(drop=True)
qatar_qualifying[['DriverName','GridPos','WinProb']]
```
